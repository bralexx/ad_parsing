```python
import pandas as pd
import json
import re
```

```python
manual = pd.read_csv('data/manual_test_100.csv', index_col=0, converters={'json': json.loads}).sample(frac=1, random_state=42)
```

```python
df = manual[['Text']].copy()
df['GPT-4'] = manual['json'].apply(lambda js: json.dumps(js, indent=4, ensure_ascii=False))
```

```python
def deduplicate(js):
  norm = [{key: value.strip().lower() for key, value in bundle.items()} for bundle in js]
  unique_bundles = []
  unique_norm = []
  for i in range(len(norm)):
    if norm[i] not in unique_norm:
      unique_bundles.append(js[i])
      unique_norm.append(norm[i])
  return unique_bundles

def deduplicate_s(s):
  js = json.loads(s)
  js = deduplicate(js)
  return json.dumps(js, indent=4, ensure_ascii=False)
```

```python
df['fred-1.7B'] = pd.read_csv('fredT5-1.7B/manual_test_outputs.csv', index_col=0)['json'].apply(deduplicate_s)
df['fred-1.7B_cb'] = ''
df['fred-1.7B_label'] = ''

df['ruT5-large-fp32'] = pd.read_csv('ruT5-large/manual_test_outputs_fp32.csv', index_col=0)['json'].apply(deduplicate_s)
df['ruT5-large-fp32_cb'] = ''
df['ruT5-large-fp32_label'] = ''

df['ruT5-large2'] = pd.read_csv('ruT5-large/manual_test_outputs_default.csv', index_col=0)['json']
df['ruT5-large2_cb'] = ''
df['ruT5-large2_label'] = ''

df['ruT5-large-sep'] = pd.read_csv('ruT5-large/manual_test_outputs_sep.csv', index_col=0)['json']
df['ruT5-large-sep_cb'] = ''
df['ruT5-large-sep_label'] = ''
df['ruT5-base'] = pd.read_csv('ruT5-base/manual_test_outputs.csv', index_col=0)['json']
df['ruT5-base_cb'] = ''
df['ruT5-base_label'] = ''
df['fredT5-large2'] = pd.read_csv('fredT5-large/manual_test_outputs.csv', index_col=0)['json']
df['fredT5-large2_cb'] = ''
df['fredT5-large2_label'] = ''

df['ruT5-small'] = pd.read_csv('ruT5-small/manual_test_outputs.csv', index_col=0)['json']
df['ruT5-small_cb'] = ''
df['ruT5-small_label'] = ''
df['fredT5-distill-95m'] = pd.read_csv('fredT5-distilled/manual_test_outputs.csv', index_col=0)['json']
df['fredT5-distill_cb'] = ''
df['fredT5-distill_label'] = ''
df['fredT5-large'] = pd.read_csv('fredT5-large/manual_test_outputs.csv', index_col=0)['json']
df['fredT5-large_cb'] = ''
df['fredT5-large_label'] = ''
```

```python
baseline_labeling = pd.read_csv('metrics_research/manual_test_labeled.csv')
joined = df.join(baseline_labeling.set_index('Text'), on='Text', how='inner')
df['GPT-4'] = joined['GptOutput']
df['GPT-4_cb'] = joined['correct bundles gpt']
df['GPT-4_label'] = joined['label gpt']
df['ruT5-large'] = joined['Output']
df['ruT5-large_cb'] = joined['correct bundles']
df['ruT5-large_label'] = joined['label']
df['tb'] = joined['true bundles']
```

```python
models = ['ruT5-small', 'fredT5-distill', 'GPT-4', 'ruT5-large', 'ruT5-large-sep', 'ruT5-base', 'fredT5-large2', 'ruT5-large2', 'ruT5-large-fp32', 'fred-1.7B']
df_cmp = pd.DataFrame(index=models)
df_cmp
```

```python
for model in models:
  df[model + '_EB_corr'] = df[model + '_label'].apply(lambda s: 0 if 'EB' not in s else int(re.findall(r'EB(.*?)(,|$)', s)[0][0]))
```

```python
df_cmp['sample_precision'] = {model: (df[model + '_cb'] / df[model].apply(json.loads).apply(len)).mean() for model in models}
df_cmp['sample_recall'] = {model: ((df[model + '_cb'] - df[model + '_EB_corr']) / df['tb']).mean() for model in models}
df_cmp['cb'] = {model: df[model + '_cb'].sum() for model in models}
df_cmp['EB_corr'] = {model: df[model + '_label'].apply(lambda s: 0 if 'EB' not in s else int(re.findall(r'EB(.*?)(,|$)', s)[0][0])).sum() for model in models}
df_cmp['cub'] = df_cmp.cb - df_cmp['EB_corr']
df_cmp['pred'] = {model: df[model].apply(lambda s: len(json.loads(s))).sum() for model in models}
df_cmp['tb'] = {model: df['tb'].sum() for model in models}
df_cmp['not_BO'] = 89
labels = ['BO', 'FM', 'EB', 'MB', 'AC']
for label in labels:
  df_cmp[label] = {model: df[model + '_label'].apply(lambda s: label in s).sum() for model in models}

# df_cmp['precision'] = (df_cmp.cb - df_cmp.EB_corr) / (df_cmp.pred - df_cmp.EB_corr)
df_cmp['precision'] = (df_cmp.cb) / (df_cmp.pred)

df_cmp['recall'] = (df_cmp.cb - df_cmp.EB_corr) / (df_cmp.tb)
```
