```python
# Define a metaclass
class CustomMeta(type):
    def __new__(cls, name, bases, dct):
        # Add a new method to the class
        dct['custom_method'] = lambda self: print(f"Custom method called by {self}")
        return super().__new__(cls, name, bases, dct)

# Use the metaclass to create a class
class MyClass(metaclass=CustomMeta):
    def __init__(self, value):
        self.value = value

    def display(self):
        print(f"Value: {self.value}")

# Create an instance of the class
obj = MyClass(42)

# Call the custom method added by the metaclass
obj.custom_method()

# Call the original method defined in the class
obj.display()
```

```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        answer = ListNode()
        current = answer
        while list1.next is not None or list2.next is not None:
            if list1.next is None:
                current.next = list2
                list2 = list2.next
            elif list2.next is None:
                current.next = list1
                list1 = list1.next
            else:
                if list1.val < list2.val:
                    current.next = list1
                    list1 = list1.next
                else:
                    current.next = list2
                    list2 = list2.next
            current = current.next
        return answer.next    
```
