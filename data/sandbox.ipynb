```python
bigger_sample = stupid_request('select * from posts limit 100000, 200000')
unique_attachment = [row for row in bigger_sample if any([obj['type'] != 'photo' for obj in json.loads(row[-1])['attachments']])]
len(unique_attachment)
```

```python
np.unique([obj['type'] for row in unique_attachment for obj in json.loads(row[-1])['attachments']])
```

```python
count = sum(["\"type\": \"link\"" in row[-1] for row in unique_attachment])
print(count)
print(count / len(unique_attachment))
print(len(bigger_sample))
print(count / len(bigger_sample))
```

```python
query = "select count(*) from (select * from posts limit 100000, 200000) as subposts where other like \"%\\\"type\\\": \\\"link\\\"%\""
print(query)
stupid_request(query)
```
